```python
#preparing the data for training
X=pd.DataFrame(np.c_[df1['LSTAT'], df1['RM']], columns=['LSTAT','RM'])
Y=df1['MEDV']
```

```python
# we train our model with 80% accuracy and test the remaining 20%. we do this to assess the model's performance on unseen data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#using Linear Regression to train our model on both training and test sets
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training test
y_train_predict= lin_model.predict(X_train)
rmse=(np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2= r2_score(Y_train, y_train_predict)

print('The model performance for training test')
print('.......................................')

print('RSME IS {}'.format(rmse))
print('R2 score is {}'.format(r2))
print('\n')

#model evaluation for testing


y_test_predict= lin_model.predict(X_test)
rmse=(np.sqrt(mean_squared_error(Y_test,y_test_predict)))

print('The model performance for testing test')
print('.......................................')

print('RSME IS {}'.format(rmse))
print('R2 score is {}'.format(r2))

    
```

```python
#pipeline helps to prevent data leakage in our test harness by ensuring that data preparation like standardization
#is contrained yo each fold of our cross validation procedure
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# StandardScaler is for removing the mean and scaling to unit variance
```

```python
pipe= Pipeline([
    ('scale',StandardScaler()),
    ('pca',PCA(n_components=5)),
    ('svr',SVR()),
])
```

```python
pipe.fit(x_train, y_train)
pipe.score(x_test, y_test)
```
